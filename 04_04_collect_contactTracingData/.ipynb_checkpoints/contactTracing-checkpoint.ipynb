```python
files = ["04_03_exp1_data.txt","04_03_exp2_data.txt","04_03_exp3_data.txt","04_03_exp4_data.txt"]
```

```python
def store_data(filename):

    origin = []
    dot_3 = []
    dot_4 = []
    s= set()
    with open("./data/"+filename,"r") as f:
        for line in f:
            data = eval(line.rstrip())
            if data["Patient_first_name"]=="192.168.4.2":
                origin.append(data)
            elif data["Patient_first_name"]=="192.168.4.3":
                dot_3.append(data)
            elif data["Patient_first_name"]=="192.168.4.4":
                dot_4.append(data)
    return origin, dot_3, dot_4

def data_from_origin(origin):
    l_dot_3 = []
    l_dot_4 = []
    for each in origin:
        l_dot_3.append(each['USR']['Device0'])
        l_dot_4.append(each['USR']['Device2'])
    return l_dot_3, l_dot_4
```

# exp1
- .3 <=> device0  =>>> 3-6
- .4 <=> device2  =>>> 6+
- origin: .2 <=> device1

# visualization

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
import seaborn as sns
```

```python
def plot_2_distribution(filename,expn, legend1,legend2):
    # legend1 for dot_3
    
   
    origin, dot_3, dot_4 =  store_data(filename)
    l_dot_3,l_dot_4 = data_from_origin(origin)
    l1,l2 = l_dot_3,l_dot_4
    
    from scipy.stats import norm
    fig, ax = plt.subplots(figsize=(14, 5))

    sns.distplot(l1, hist=True, fit=norm, kde=False, 
                 bins=max(l1)-min(l1), color = 'darkblue', 
                 hist_kws={'edgecolor':'black'},
                 kde_kws={'linewidth': 4},label=legend1)
    sns.distplot(l2, hist=True, fit=norm, kde=False, 
                 bins=max(l2)-min(l2), color = 'green', 
                 hist_kws={'edgecolor':'yellow'},
                 kde_kws={'linewidth': 4},label=legend2)
    plt.legend(loc=2, prop={'size': 20})
    plt.title(f"{expn}: maximum likelihood gaussian distribution fit", fontsize=20)
    fig.show()
```

# Descriptions:

- exp1: haohua：.3 3-6 weisheng：.4 6+
- exp2: haohua：.3 0-3 weisheng：.4 6+
- exp3: weisheng：.3 3-6 haohua：.4 6+
- exp4: weisheng：.3 0-3 haohua：.4 6+

```python
plot_2_distribution(files[0], "exp1","3~6","6+")
```

```python
origin, dot_3, dot_4 =  store_data(files[0])
l_dot_3_3_6,l_dot_4_6plus1 = data_from_origin(origin)
origin, dot_3, dot_4 =  store_data(files[1])
l_dot_3_0_3,l_dot_4_6plus2 = data_from_origin(origin)
l_dot_4_6plus = l_dot_4_6plus1 + l_dot_4_6plus2

from scipy.stats import norm
fig, ax = plt.subplots(figsize=(14, 5))
l1,l2,l3 = l_dot_3_0_3,l_dot_3_3_6,l_dot_4_6plus
legend1,legend2,legend3 = "0-3","3-6","6+" 
sns.distplot(l1, hist=True, fit=norm, kde=False, 
             bins=max(l1)-min(l1), color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4},label=legend1)
sns.distplot(l2, hist=True, fit=norm, kde=False, 
             bins=max(l2)-min(l2), color = 'green', 
             hist_kws={'edgecolor':'yellow'},
             kde_kws={'linewidth': 4},label=legend2)
sns.distplot(l3, hist=True, fit=norm, kde=False, 
             bins=max(l3)-min(l3), color = 'red', 
             hist_kws={'edgecolor':'yellow'},
             kde_kws={'linewidth': 4},label=legend3)
plt.legend(loc=2, prop={'size': 20})
plt.title(f"exp1+2: maximum likelihood gaussian distribution fit", fontsize=20)
fig.show()
```
